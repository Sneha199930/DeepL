```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# independent
x = np.array([-1,0,1,2,3,4])
```

```python
x
```

```python
# dependent
y = np.array([-3,-1,1,3,5,7])
```

```python
y
```

```python
x.shape
```

```python
x = x.reshape(-1,1)
```

```python
x.shape
```

```python
x
```

```python
plt.scatter(x,y)
plt.xlabel("Independent")
plt.ylabel("Dependent")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(x,y)
```

```python
lin_model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lin_model.predict(x),color='red')
plt.xlabel("Independent")
plt.ylabel("Dependent")
plt.show()
```

```python
# lasso
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=0.8)
```

```python
lasso_model.fit(x,y)
```

```python
lasso_model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lasso_model.predict(x),color='red')
plt.xlabel("Independent")
plt.ylabel("Dependent")
plt.show()
```

```python
# Ridge
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=4.5)
```

```python
ridge_model.fit(x,y)
```

```python
ridge_model.score(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,ridge_model.predict(x),color='red')
plt.xlabel("Independent")
plt.ylabel("Dependent")
plt.show()
```
